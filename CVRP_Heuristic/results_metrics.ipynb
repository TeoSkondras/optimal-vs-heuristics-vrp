```python
#on comparison txt we have in each line the isntance name and our ratio of the best known solution to our solution, find the average
#open the file
with open("comparison.txt", "r", encoding="utf-16") as f:
    lines = f.read().splitlines()

sum = 0
count = 0
for line in lines:
    parts = line.split()
    if len(parts) == 2:
        print(parts[0], parts[1])
        ratio = float(parts[1])
        sum += ratio
        count += 1

print("Average ratio:", sum / count)
```
